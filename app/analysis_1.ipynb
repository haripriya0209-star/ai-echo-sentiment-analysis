```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv(r"D:\Sentimental analysis\cleaned_reviews_dataset.csv")
```

```python
df.dropna(subset=['cleaned_review'], inplace=True)
```

```python
# Split FIRST (Keep Test data pure/unbalanced to reflect real world)
X_train, X_test, y_train, y_test = train_test_split(
    df['cleaned_review'], 
    df['label_encoded'], 
    test_size=0.2, 
    random_state=42
)
```

```python
print(f"Training on {len(X_train)} samples")
print(f"Testing on {len(X_test)} samples")
```

```python
tfidf = TfidfVectorizer(max_features=5000,ngram_range=(1,2))
```

```python
# FIT only on Train, TRANSFORM on both
X_train_vec = tfidf.fit_transform(X_train) # Learns vocabulary here
X_test_vec  = tfidf.transform(X_test)
```

```python
print(f"Train Shape: {X_train_vec.shape}")
print(f"Test Shape:  {X_test_vec.shape}")
```

```python
model = LogisticRegression(class_weight='balanced',max_iter=1000)
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)
print("\n--- Results ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
print("Training Random Forest...")
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_vec, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test_vec)
# 4. Results
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.svm import LinearSVC
print("Training SVM Model...")
svm_model = LinearSVC(C=1.0, class_weight='balanced', max_iter=10000, random_state=42)
svm_model.fit(X_train_vec, y_train)
```

```python
y_pred_svm = svm_model.predict(X_test_vec)
```

```python
# 4. Results
print(f"Accuracy: {accuracy_score(y_test, y_pred_svm):.4f}")
print(classification_report(y_test, y_pred_svm))
```
